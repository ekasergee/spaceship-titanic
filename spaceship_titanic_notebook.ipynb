### Логистическая регрессия

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
```

```python
def print_scores(model, X_train, y_train, X_valid, y_valid):
  print("VALID")
  print(classification_report(y_valid, model.predict(X_valid)))
  print("TRAIN")
  print(classification_report(y_train, model.predict(X_train)))
  # Построим матрицу ошибок
  plot_confusion_matrix(model, X_valid, y_valid)
  plt.show()
```

```python
def fit_logistic_regression(X_train, y_train, X_valid, y_valid):
  # создадим и обучим модель логистической регрессии
  model_log = LogisticRegression()

  model_log.fit(X_train, y_train)

  # выполним предсказания и посмотрим результаты метрик
  print_scores(model_log, X_train, y_train, X_valid, y_valid)
  return model_log
```

```python
X = df_train_mod_ohe.copy()
y = df_train['Transported']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
# выполним нормализацию
scaler = StandardScaler()
scaler.fit(X_train, y_train)

X_train_st = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
X_valid_st  = pd.DataFrame(scaler.transform(X_valid), index=X_valid.index, columns=X_valid.columns)

# создадим и обучим модель
model_log = fit_logistic_regression(X_train_st, y_train, X_valid_st, y_valid)

logist_regr_acc = metrics.accuracy_score(y_valid, model_log.predict(X_valid_st))
logist_regr_prec = metrics.precision_score(y_valid, model_log.predict(X_valid_st))
logist_regr_recall = metrics.recall_score(y_valid, model_log.predict(X_valid_st))
logist_regr_f1 = metrics.f1_score(y_valid, model_log.predict(X_valid_st))
```

### KNN

```python
def fit_KNN(X_train, y_train, X_valid, y_valid):
  model_knn = KNeighborsClassifier()
  model_knn.fit(X_train, y_train)
  print_scores(model_knn, X_train, y_train, X_valid, y_valid)
  return model_knn
```

```python
# создадим и обучим модель
model_knn = fit_KNN(X_train, y_train, X_valid, y_valid)
```

```python
knn_acc = metrics.accuracy_score(y_valid, model_knn.predict(X_valid))
knn_prec = metrics.precision_score(y_valid, model_knn.predict(X_valid))
knn_recall = metrics.recall_score(y_valid, model_knn.predict(X_valid))
knn_f1 = metrics.f1_score(y_valid, model_knn.predict(X_valid))
```

### LGBM

```python
import lightgbm as lgb
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
def fit_grid_LGBM(X_train, y_train, X_valid, y_valid):
  grid_LGBM = RandomizedSearchCV(lgb.LGBMClassifier(),
                               param_distributions = {
                                  'max_depth': range(2, 7),
                                  'learning_rate': np.linspace(0.001, 0.3, 100),
                              },
                               scoring='accuracy',
                               cv=3,
                               n_jobs=-1,
                               return_train_score=True,
                               n_iter=100,
                              )

  grid_LGBM.fit(X_train, y_train)
  print(f"Лучшие параметры: {grid_LGBM.best_params_}")
  print_scores(grid_LGBM.best_estimator_, X_train, y_train, X_valid, y_valid)
  return grid_LGBM.best_estimator_
```

```python
best_LGBM = fit_grid_LGBM(X_train, y_train, X_valid, y_valid)
```

```python
lgb_acc = metrics.accuracy_score(y_valid, best_LGBM.predict(X_valid))
lgb_prec = metrics.precision_score(y_valid, best_LGBM.predict(X_valid))
lgb_recall = metrics.recall_score(y_valid, best_LGBM.predict(X_valid))
lgb_f1 = metrics.f1_score(y_valid, best_LGBM.predict(X_valid))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
def fit_grid_RFC(X_train, y_train, X_valid, y_valid):
  grid_RFC = RandomizedSearchCV(
            RandomForestClassifier(),
            param_distributions = {
                'n_estimators': range(100, 200),
                'max_depth': range(6, 20),
                'min_samples_split': range(2, 9),
                'min_samples_leaf': range(1, 10),
                'max_leaf_nodes': range(2, 400)
            },
            scoring = 'accuracy',
            cv = 3,
            n_jobs = -1,
            return_train_score = True,
            n_iter=100,
        )

  grid_RFC.fit(X_train, y_train)
  print(f"Лучшие параметры: {grid_RFC.best_params_}")
  print_scores(grid_RFC.best_estimator_, X_train, y_train, X_valid, y_valid)
  return grid_RFC.best_estimator_
```

```python
best_RFC = fit_grid_RFC(X_train, y_train, X_valid, y_valid)
```

```python
rfc_acc = metrics.accuracy_score(y_valid, best_RFC.predict(X_valid))
rfc_prec = metrics.precision_score(y_valid, best_RFC.predict(X_valid))
rfc_recall = metrics.recall_score(y_valid, best_RFC.predict(X_valid))
rfc_f1 = metrics.f1_score(y_valid, best_RFC.predict(X_valid))
```

Соберем полученные метрики для рассмотренных моделей в таблицу

```python
def show_all_scores(X_valid, y_valid, **models):
  model_scores = {}
  for name, model in models.items():
    model_scores[name] = [metrics.accuracy_score(y_valid, model.predict(X_valid)),
                          metrics.precision_score(y_valid, model.predict(X_valid)),
                          metrics.recall_score(y_valid, model.predict(X_valid)),
                          metrics.f1_score(y_valid, model.predict(X_valid))]

  df_scores = pd.DataFrame(model_scores,
                         index=['Accuracy', 'Precision', 'Recall', 'F1'])
  print(df_scores)
```

```python
%%time
# Воспользуемся поиском важности перемешиванием, для лучшей модели LGBM

result_LGBM = permutation_importance(best_LGBM, X_valid, y_valid,
                                     n_repeats=30,
                                     random_state=0)

# Воспользуемся поиском важности перемешиванием, для лучшей модели CatBoost

result_RFC = permutation_importance(best_RFC, X_valid, y_valid,
                                     n_repeats=30,
                                     random_state=0)
```

```python
X_valid.columns[(result_LGBM['importances_mean'] > 0) | (result_RFC['importances_mean'] > 0)]
```

```python
# признаки, оказавшиеся неважными
X_valid.columns[(result_LGBM['importances_mean'] <= 0) & (result_RFC['importances_mean'] <= 0)]
```

Рассмотрим подробнее признаки, оказавшиеся неважными.

Добавленный признак alone_at_cabin оказался неважным. При предварительном анализе выглядело, что у людей путешествующих в одиночку в каюте ниже шанс попасть в другое измерение, но данная гипотеза не подтвердилась.

Признак Destination_PSO J318.5-22 оказался неважным признаком. По предварительному анализу также получалось, что нет зависимости между этим прикнаком и таргетным.

Признак, обозначающий палубу T оказался также неважным. При предварительном анализе было видно, что в случае некоторых палуб присутствует зависимость с таргетным признаком, для некоторых этой зависимости нет. Как показала проверка важности признаков перемешиванием для этих палуб зависимость с таргетным признаком отсутствует.
